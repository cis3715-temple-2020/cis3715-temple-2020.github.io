# About Data Science

**What is data science**. Data science refers to obtaining, processing, getting insights about potentially large quantities or messy data, and communicating those insights to others. To be successful in data science takes a wide range of technical skills, including programming, algorithmic thinking, statistical thinking, and math. Good data scienctists also have a range of soft skills, including perseverance, curiosity, and ability to work in teams with diverse expertise. The last, but not least, having talent for design and communication with non-experts can be a great plus.

**Why learn data science**. Data science is the job of 21st century. As we are creating more and more data as part of our daily lives, as scientists are developing better techniques to collect data, and businesses and governments are starting to understand the potential of big data, there is an increased need for people able to handle, visualize, analyze, and obtain insights from data. This is already reflected in the job market, where *data scientist* jobs are starting to outnumber some traditional jobs for computer scientists.

**Where is data coming from?** Data for data science could be obtained in multiple ways. For example, it might be provided by your client as a spreadsheet or a database. Sometimes, your client or yourself have a data science project in mind, but do not have the data ready. Luckily, there are many data sets in a decent format already available for playing:
- Kaggle
- UCI
- google search (it is amazing how much data is lying around)

Very often, the data exists on the web, but needs to be scraped and put in the right shape through some programming, before you can do any data analysis. Python is a great tool to do it.
